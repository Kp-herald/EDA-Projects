## converting reviews into int

```python
df_copy["Reviews"]=df_copy["Reviews"].astype(int)
```

```python
df_copy.info()
```

```python
df_copy.head(2)
```

```python
df_copy.describe()
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"]=df_copy["Size"].str.replace("M","000")
df_copy["Size"]=df_copy["Size"].str.replace("k","")
df_copy["Size"]=df_copy["Size"].replace("Varies with device",np.nan)
df_copy["Size"]=df_copy["Size"].replace("",np.nan)
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"]=df_copy["Size"].astype(float)
```

```python
df_copy.info()
```

```python
df_copy["Installs"].unique()
```

```python
df_copy.head(2)
```

```python
df["Price"].unique()
```

```python
items_remove=["+",",","$"]
cols_remove=["Installs","Price"]
for i in items_remove:
    for cols in cols_remove:
        df_copy[cols]=df_copy[cols].str.replace(i,"")
```

```python
df_copy["Price"].unique()
```

```python
"Everyone" in df_copy["Price"].values
```

```python
df_copy["Installs"].unique()
```

```python
df_copy["Installs"]=df_copy["Installs"].astype(int)
df_copy["Price"]=df_copy["Price"].astype(float)
```

```python
df_copy.info()
```

```python
df_copy.head(1)
```

```python
df_copy["Last Updated"].unique()
```

```python
df_copy["Last Updated"]=pd.to_datetime(df_copy["Last Updated"])
```

```python
df_copy.head(1)
```

```python
df_copy["Day"]=df_copy["Last Updated"].dt.day
df_copy["Month"]=df_copy["Last Updated"].dt.month
df_copy["Year"]=df_copy["Last Updated"].dt.year
```

```python
df_copy.head(1)
```

```python
df_copy.drop("Last Updated",axis=1,inplace=True)
```

```python
df_copy.head(1)
```

```python
df_copy.to_csv("data/googlecleared.csv")
```

```python
numerical_feature=[feature for feature in df_copy.columns if df_copy[feature].dtype !='str']
categorial_feature=[feature for feature in df_copy.columns if df_copy[feature].dtype =='str']

print(f"number are: {len(numerical_feature)} that are {numerical_feature}")
print(f"number are: {len(categorial_feature)} that are {categorial_feature}")
```

```python
# numerical graphical representation
plt.figure(figsize=(10,10))
plt.suptitle("analysis of numerical features")
for i in range(0,len(numerical_feature)):
    plt.subplot(3,3,i+1)
    sb.kdeplot(x=df_copy[numerical_feature[i]],fill=True,color="g",ec="black")
    plt.tight_layout()
    plt.xlabel(numerical_feature[i])
    
```

# observations
-Rating and Year are right skewed whereas Reviews,Size,Installs,Price are left skewed

```python
# categorial representation
plt.figure(figsize=(30,25))
categorical_feature = ['Type','Content Rating','Android Ver']
plt.suptitle("Analysis of categorial feature",fontsize=40,alpha=1,y=1,fontweight="bold")
for i in range(len(categorical_feature)):
    plt.subplot(2,2,i+1)
    sb.countplot(x=df_copy[categorical_feature[i]],  hue=df_copy[categorical_feature[i]], palette="Set3")
    plt.xticks(rotation=45)
    plt.tight_layout()
    
```

# Category that has Largest number Installations

```python
df_copy.groupby("Category")["Installs"].sum().sort_values(ascending=False).plot.pie(figsize=(30,20))
```

#### observation:Game has the most installations

```python
apps = ['GAME','COMMUNICATION','PRODUCTIVITY','SOCIAL']
plt.figure(figsize=(20,15))

for i,cat in enumerate(apps):
    plt.subplot(2,2,i+1)

    top_apps = df_copy[df_copy['Category']==cat].sort_values('Installs',ascending=False).head(5)

    sb.barplot(data=top_apps, x='Installs', y='App',hue=df['Category'],palette='Set2')
    plt.title(cat)
    plt.xlabel("Installs")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()
```

## Top 10 Apps that got 5 rated in google playstore

```python
toprating_apps = df_copy[df_copy['Rating'] == 5.0]
top_apps = toprating_apps.sort_values("Installs", ascending=False).head(10)
sb.barplot(data=top_apps, x="Installs", y="App", hue="App",palette="hls")
plt.title("Top 10 Apps with Rating 5.0")
plt.show()
```
